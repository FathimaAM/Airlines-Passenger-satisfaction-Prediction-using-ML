**CONVERTING CATEGORICAL VALUES TO NUMERICAL VALUES**

```python
import numpy as np

new_df = updt_df.copy()

new_df['satisfaction'] = np.where(new_df['satisfaction'] == 'satisfied', 0, 1)

new_df['Gender'] = np.where(new_df['Gender'] == 'Male', 0, 1)

new_df['type_of_travel'] = np.where(new_df['type_of_travel'] == 'Personal Travel', 0, 1)

new_df['customer_class'] = np.select(
    [
        new_df['customer_class'] == 'Eco',
        new_df['customer_class'] == 'Eco Plus',
        new_df['customer_class'] == 'Business'
    ],
    [0, 1, 2],
    default=np.nan
)

new_df['customer_type'] = np.where(new_df['customer_type'] == 'Loyal Customer', 0, 1)
```

```python
new_df.head()
```

**FINDING AND PLOTTING IMPORTANT FEATURES**

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import pandas as pd

# Assume 'satisfaction' is the target variable in new_df
X = new_df.drop('satisfaction', axis=1)  # Features
y = new_df['satisfaction']               # Target variable

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Train Random Forest model
clf = RandomForestClassifier(n_estimators=100, random_state=0)
clf.fit(X_train, y_train)

# Get feature importances
importances = clf.feature_importances_

# Create a DataFrame to display the feature importances
importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

print(importance_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting the feature importances
plt.figure(figsize=(10, 8))
sns.barplot(x='Importance', y='Feature', data=importance_df)

# Adding titles and labels
plt.title('Important feature selection')
plt.xlabel('Importance')
plt.ylabel('Feature')

# Show the plot
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(X)
X_scaled=scaler.transform(X)
X_scaled
```
